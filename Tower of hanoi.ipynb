## solving the tower of hanoi problem 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

def hanoi(a,b,c,n):
    global count
#     Base case
    if n==1:
        disk = a.pop()
        c.append(disk)
        count +=1
    else:
        hanoi(a,c,b,n-1)
        hanoi(a,b,c,1)
        hanoi(b,a,c,n-1)
    return count
```
