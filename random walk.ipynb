```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
```

```python
plt.rc('figure',figsize=(12,6))
```

```python
def random_walk(n,m):
    X,Y= [0], [0]
    for i in range (1,n):
        dx, dy = choice([(1,0),(-1,0),(0,1),(0,-1)])
        y= Y[i-1]+dy
        x= X[i-1]+dx
        X.append(x)
        Y.append(y)
    if m == 0:
        return x,y
    return X,Y,x,y
```

```python
X,Y,x,y = random_walk(99)
```

```python
print(x,y)
```

```python
plt.plot(X,Y,'ko--')
plt.plot([0,x],[0,y])
plt.title('Our Plot')
plt.xlabel('This is the horizantal axis')
plt.ylabel('Vertical axis')
plt.show()
```

## what is the longest random walk you can take so that on average you end up four blocks or fewer from home?

### you will be able to return home walking without paying for transport if you end up four blocks or fewer to home, we will calculate the probability of each random walk with different number of walks using monte carlo method, and from all the walks with probability higher than 0.5 we will choose the one with the higher number of walks

```python
def random_walk_2(n):
    x,y = 0,0
    for i in range (n):
        dx, dy = choice([(1,0),(-1,0),(0,1),(0,-1)])
        x+= dx
        y+= dy
    return x,y
```

```python
number_of_trials = 30000
# we will try from 1 step to 30
l =[]
for walk_length in range(1,31):
    no_transport = 0
    for j in range(number_of_trials):
        x, y = random_walk_2(walk_length)
        if abs(x)+abs(y) <=4:
            no_transport+=1
    no_transport_percentage = no_transport/number_of_trials
    print(f'walk length of {walk_length} % of no transport = {no_transport_percentage}')
    l.append(no_transport_percentage)
plt.plot(range(1,31),l)
```

```python
plt.plot(range(1,31),l)
plt.plot(range(1,31),[0.5]*30)
plt.title('Monte carlo')
plt.xlabel('number of walks')
plt.ylabel('probability that we will end up 4 blocks or less from home')
plt.show()
```
